```python
data = pd.read_csv("data/한국전력공사_전기차 충전소 충전량_20220330.csv")
data.head()
```

```python
# 필요없는 컬럼 삭제
data = data.drop(["본부","사업소","충전기"], axis=1)
data
```

```python
data.info()
```

```python
data = data[(data['충전용량']==7) | (data['충전용량']==50)]
data
```

```python
data = data[(data['충전량']<=80)]
data
```

```python
data.info()
```

```python
data = data[~data['주소'].str.contains("null", na=False, case=False)]
```

```python
data = data.dropna(axis=0)
data
```

```python
data.info()
```

```python
data["충전시간(1)"] = (((data["충전시간"] * 60) + data["충전분"]) / 60).round(2)
data.head()
```

```python
data = data[(data['충전시간(1)']>=0.08) & (data['충전시간(1)']<=24)]
data
```

```python
data = data.drop(["충전시간","충전분","충전종료일자"], axis=1)
data
```

```python
data.rename(columns={"충전시간(1)": "충전시간"}, inplace=True)
```

```python
data.info()
```

```python
data = data[~data['주소'].str.contains("제주", na=False, case=False)]

data
```

```python
data.info()
```

```python
# 실충전시간 컬럼 생성 충전량/충전용량, 미사용시간 컬럼생성 충전시간-실충전시간
data["실충전시간"] = (data["충전량"] / data["충전용량"]).round(2)
data.head()
```

```python
data["미사용시간"] = (data["충전시간"] - data["실충전시간"]).round(2)
data.head()
```

```python
# 중간 저장
data.to_csv("data/Pre_Charging.csv", encoding="utf-8", index=False)
```
