```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm # 폰트 설정
font_name = fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc("font", family=font_name)

import matplotlib as mlp # 음수 설정
mlp.rcParams["axes.unicode_minus"] = False

cartype = pd.read_csv("../data/자동차+등록현황(연료별)_20221025141634.csv")
# cartype.head()
# cartype.info()

# cartype["2017. 01"].str.replace("-","0")
cartype = cartype.replace("-","0")
# cartype1.isnull().sum()
cartype =cartype.fillna(0)
# cartype.head(50)
# cartype.info()
del cartype["차종별(1)"]
del cartype["연료별(1)"]
cartype = cartype.set_index(["차종별(2)","연료별(2)"])
cartype = cartype.astype("float64")
# cartype.info()
cartype.head(50)
```

```python
cartype1 = cartype
cartype1 = cartype1.drop(["승용차","승합차","화물차","특수차"])
cartype1 = cartype1.reset_index(["차종별(2)","연료별(2)"])
del cartype1["차종별(2)"]
cartype1.rename(columns={'연료별(2)':'연료타입'}, inplace=True)
cartype1 = cartype1.set_index(["연료타입"])
cartype1.rename(index={'소계':'종합'}, inplace=True)
cartype1 = cartype1.drop(["CNG","기타","수소"])
cartype1
```

```python
cartype1['소계'] = cartype1['2021. 12']
cartype1['2017년 증감율'] = (cartype1['2017. 12'] - cartype1['2017. 01']) / cartype1['2017. 01'] * 100
cartype1['2018년 증감율'] = (cartype1['2018. 12'] - cartype1['2018. 01']) / cartype1['2018. 01'] * 100
cartype1['2019년 증감율'] = (cartype1['2019. 12'] - cartype1['2019. 01']) / cartype1['2019. 01'] * 100
cartype1['2020년 증감율'] = (cartype1['2020. 12'] - cartype1['2020. 01']) / cartype1['2020. 01'] * 100
cartype1['2021년 증감율'] = (cartype1['2021. 12'] - cartype1['2021. 01']) / cartype1['2021. 01'] * 100
```

```python
print((cartype1['2021. 12'] - cartype1['2017. 01']) / cartype1['2017. 01'] * 100)
print((cartype1['2017년 증감율'] + cartype1['2018년 증감율'] + cartype1['2019년 증감율'] + cartype1['2020년 증감율']
      + cartype1['2021년 증감율']) / 5)
```
