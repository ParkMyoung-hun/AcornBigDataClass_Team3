```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from tqdm.notebook import tqdm
import pandas as pd
import folium
import time 

try:
    url = "http://data.ex.co.kr/link/linkList?pn=1&linkId=2"
    driver = webdriver.Chrome("chromedriver")
    driver.get(url)
    x_lng = []
    y_lat = []
    station_name = []

    for page_num in tqdm(range(2,38)):
        tr = driver.find_elements(By.ID, "0")
        for i in tr:
            x_lng.append(i.find_element(By.CLASS_NAME, "XVALUE").text)
            y_lat.append(i.find_element(By.CLASS_NAME, "YVALUE").text)
            station_name.append(i.find_element(By.CLASS_NAME, "UNITNAME").text)

        page = driver.find_element(By.LINK_TEXT, str(page_num))
        page.click()
        time.sleep(1)
    
    

except Exception as e:
    print(e)
finally:
    driver.close()
    driver.quit()

stations = pd.DataFrame({"영업소명":station_name, "경도":x_lng, "위도":y_lat})
map = folium.Map(location=[37.497911, 127.027558], zoom_start=7)
for i in  tqdm(stations.index):
    marker = folium.Marker(location=[stations["위도"][i], stations["경도"][i]], 
                                   icon=folium.Icon(color="red", icon="star"), popup=stations["영업소명"][i])
    marker.add_to(map)
map
```
