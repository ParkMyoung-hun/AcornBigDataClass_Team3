### 카카오API 사용

```python
import requests, json
import time
 
def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK ff73c4d90fcdae2be25ed5ea1177ca57"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json
```

```python
get_location('인천 서구 비즈니스로 41')
```

```python
def result_location(i):
    api_json = get_location(data['주소'][i])
    address = api_json['documents'][0]['address']
    data.loc[i,'x'] = address['x']; 
    data.loc[i,'y'] = address['y']
```

```python
get_location(data['주소'][0])['documents'][0]['address']['x']
```

```python
result_location(40)
```

```python
with tqdm(total = 1000) as pbar:
    i = 0
    while i<=1000:
        try:
            result_location(i)
            i+=1
        except:
            print(i, '번째', 'time.sleep 적용합니다.')
            time.sleep(2)
            result_location(i)
            i+=1

#인천... 에러 뜸.. 주소는 인천광역시인데 카카오API는 인천이네..
```

```python
#전체데이터 돌릴 코드
with tqdm(total = len(data)) as pbar:
    i = 0
    while i<=len(data['주소']):
        try:
            result_location(i)
            i+=1
        except:
            print(i, '번째', 'time.sleep 적용합니다.')
            time.sleep(2)
            result_location(i)
            i+=1
```
