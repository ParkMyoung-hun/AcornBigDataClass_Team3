```python
EVcharging = pd.read_csv("../data/EVcharging.csv", encoding="euc-kr")
EVcharging.head(1)
```

```python
ev = pd.DataFrame({"충전소명":EVcharging["충전소명"], "시도":EVcharging["설치시도명"], "주소":EVcharging["소재지도로명주소"], "위도":EVcharging["위도"], "경도":EVcharging["경도"]})
ev.head()
```

```python
import folium
```

```python
map = folium.Map(location=[37.497911, 127.027558], zoom_start=9)

for i in ev.index:
    lat = ev["위도"][i]
    lng = ev["경도"][i]

    folium.CircleMarker(location=[lat, lng], radius=2, color="green", 
                        popup=ev["충전소명"][i]).add_to(map)
map
```
